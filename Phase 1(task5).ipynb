```python
plt.figure(figsize=(30, 25))  # Adjust width and height as needed
ax = sns.countplot(y='TOTALVALUE', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(23, 36))
a=df.groupby(['TOTALVALUE'],as_index=False)['ID'].sum().sort_values(by='ID',ascending=True)
sns.barplot(x='ID',y='TOTALVALUE',data=a)
```

```python
# Extract these columns for analysis
plt.title('Correlation Matrix Heatmap')
columns_of_interest = ['ID','TOTALVALUE']
data_subset = df[columns_of_interest]

# One-hot encode categorical variables
data_subset = pd.get_dummies(data_subset, columns=['ID','TOTALVALUE'])

# Calculate the correlation matrix
correlation_matrix = data_subset.corr()

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
```
